```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('/content/ecommerce_preparados.csv')

print('Histograma - Distribuição das Notas')
sns.histplot(df['Nota'], bins=10, kde=True)
plt.title('Distribuição das Notas dos Produtos')
plt.xlabel('Notas')
plt.ylabel('Frequência de Produtos')
plt.show()


print('Dispersão - Nota x Número de Avaliações')
sns.scatterplot(x='Nota', y='N_Avaliações', data=df)
plt.title('Dispersão: Nota x Número de Avaliações')
plt.xlabel('Nota')
plt.ylabel('Nº de Avaliações')
plt.show()

print('Mapa de Calor - Correlação entre Variáveis Numéricas')
df_num = df.select_dtypes(include=[np.number])
sns.heatmap(df_num.corr(), annot=True,fmt=".1f", cmap='coolwarm')
plt.title('Mapa de Calor: Correlação entre Variáveis')
plt.show()


print('Gráfico de Barra - Top 10 Marcas mais Frequentes')
top_marcas = df['Marca'].value_counts().head(10)
top_marcas.plot(kind='bar')
plt.title('Top 10 Marcas Mais Frequentes')
plt.ylabel('Contagem')
plt.xlabel('Marca')
plt.xticks(rotation=90)
plt.show()


print('Gráfico de Pizza - Gênero dos Produtos')
# Exemplo com a coluna 'Gênero'
valores = df['Gênero'].value_counts(normalize=True)
valores_original = df['Gênero'].value_counts()
# Filtra valores >= 2%
principais = valores[valores >= 0.02]
# Soma os outros
outros = valores[valores < 0.02].sum()
# Combina tudo
valores_final = principais.copy()
valores_final['Outros'] = outros
plt.figure(figsize=(8,8))
plt.pie(valores_final, labels=valores_final.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Marcas (agrupando <2% em "Outros")')
plt.axis('equal')
plt.show()


print('Gráfico de Densidade - Preço MinMax')
sns.kdeplot(df['Preço_MinMax'], fill=True)
plt.title('Densidade do Preço (normalizado)')
plt.xlabel('Preço Normalizado')
plt.show()


print('Gráfico de Regressão - Preço vs Nota')
sns.regplot(x='Preço', y='Nota', data=df, scatter_kws={"alpha":0.5})
plt.title('Regressão: Preço vs Nota')
plt.xlabel('Preço (Normalizado)')
plt.ylabel('Nota')
plt.show()
```
